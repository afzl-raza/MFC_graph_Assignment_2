Directed Graph:

Out-degree: Number of edges leaving the vertex.

In-degree: Number of edges entering the vertex.

Example: If a vertex X has 2 outgoing edges and 1 incoming edge:

Out-degree = 2

In-degree = 1

Degree of B = 2 (edges: B-A, B-C)

Degree of C = 2 (edges: C-A, C-B)


```python
def degree_adj_list(adj_list):
    # First compute degrees for all nodes
    degrees = {key: len(neighbors) for key, neighbors in adj_list.items()}
    
    # Sort the keys based on degree (ascending order)
    sorted_keys = sorted(degrees.keys(), key=lambda x: degrees[x])
    
    # Create a new dictionary with sorted keys and their degrees
    sorted_degrees = {key: degrees[key] for key in sorted_keys}
    
    return sorted_degrees
degree_adj_list(Adjacency_list)
```

1. Edge List (Easiest)
What it is: Just a list of connections (edges) between nodes.

Format: List of tuples (u, v) where u and v are connected.

2. Adjacency List (Most Common)
What it is: A dictionary/hashmap where each node points to its neighbors.
Format: {node: [neighbor1, neighbor2, ...]}

3. Adjacency Matrix (Best for Dense Graphs)
What it is: A 2D table (matrix) where matrix[u][v] = 1 if u and v are connected, else 0.
Format: N x N grid (where N = number of nodes)

A complete graph is a simple undirected graph where every pair of distinct vertices is connected by a unique edge.


```python
def check_complet(adj_matrix):
    n = len(adj_matrix)
    for i in range(n):
        for j in range(n):
            if i != j:
                if adj_matrix[i][j] != 1:
                    return False
    return True
Ad_matrix = [[0,1,1,1,1,1], [1,0,1,1,1,1], [1,1,0,1,1,1], [1,1,1,0,1,1], [1,1,1,1,0,1], [1,1,1,1,1,0]]
(check_complet(Ad_matrix))
```

Algorithm to Check if a Graph is Connected
Input: An adjacency list representation of a graph.
Output: True if the graph is connected, otherwise False.

Choose a starting node arbitrarily.
Initialize a visited set to track visited nodes.
Perform BFS or DFS traversal from the starting node:
Add the starting node to the visited set.
Explore all reachable nodes iteratively (BFS) or recursively (DFS).
After traversal, check if all nodes are visited:
If the size of the visited set equals the number of nodes, return True (Graph is connected).
Otherwise, return False (Graph is disconnected).

```python
def check_connected(Graph):
    visited_set = set()
    for key in Graph:
        if key not in visited_set:
            stack = [key]
            while stack:
                node = stack.pop()
                if node not in visited_set:
                    visited_set.add(node)
                    stack.extend(Graph[node])
            break
    
    return len(visited_set) == len(Graph)
```

Walk 

Definition: A sequence of vertices where each adjacent pair is connected by an edge.

Rules:

Can repeat vertices and edges.

No restrictions on revisiting.

Example:

Graph: A — B — C — D

Walk: A → B → C → B → A → D (repeats edges and vertices)

Trail
Definition: A walk where no edge is repeated.

Rules:

Vertices can be repeated, but edges cannot.

Example:

Graph: A — B — C — D

Trail: A → B → C → B → D (repeats vertex B but not edges)

Path
Definition: A walk where no vertex or edge is repeated.

Rules:

No repeats of vertices or edges.

The strictest form of traversal.

Example:

Graph: A — B — C — D

Path: A → B → C → D (no repeats)



```python
def check_graph_type(graph,sequence):
    visited_edge= set()
    visited_node= set()
    for i in range(len(sequence)-1):
        u= sequence[i]
        visited_node.add(u)
        v= sequence[i+1]
        if v not in graph[u] and u not in graph[v]:
            return "not a walk"
        edge = tuple(sorted(v,u))
        visited_edge.add(edge)
        visited_node.add(u)
    visited_node.add(sequence[-1])
    is_path= len(visited_node)==len(sequence)
    is_trail= len(visited_edge)==len(sequence)-1
    if is_path:
        return "path"
    elif is_trail:
        return "trail"
    else:
        return "walk"
```

```python
def is_tree(graph):
    if not graph:
        return True  # Empty graph is a tree
    
    n = len(graph)
    m = sum(len(adj) for adj in graph.values()) // 2  # Edge count for undirected graph
    
    if m != n - 1:
        return False
    
    visited = set()
    # Simpler way to get first node:
    first_node = list(graph.keys())[0]  # Get first key from dictionary
    stack = [first_node]
    
    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            stack.extend(neighbor for neighbor in graph[node] if neighbor not in visited)
    
    return len(visited) == n
# A tree (connected, no cycles)
tree_graph = {
    0: [1, 2],
    1: [0],
    2: [0, 3],
    3: [2]
}

# Not a tree (cycle)
non_tree_graph = {
    0: [1, 2],
    1: [0, 2],
    2: [0, 1]
}

print(is_tree(tree_graph))      # Output: True
print(is_tree(non_tree_graph))  # Output: False
```

```python
def find_spanning_tree(graph):
    if not graph:
        return {}
    
    spanning_tree = {node: [] for node in graph}
    visited = set()
    
    # Get first node in a simpler way:
    first_node = list(graph)[0]  # Convert dict to list of keys and take first one
    stack = [first_node]
    visited.add(first_node)
    
    while stack:
        node = stack.pop()
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                spanning_tree[node].append(neighbor)
                spanning_tree[neighbor].append(node)  # For undirected graph
                stack.append(neighbor)
    
    return spanning_tree
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['B']
}
list(graph)[0]
```

```python
def count_leaves(graph):
    if not graph:
        return 0
    
    leaves = 0
    for node in graph:
        if len(graph[node]) == 1:  # Leaf condition
            leaves += 1
    
    # Handle single-node tree
    return leaves if len(graph) > 1 else 1
tree = {
    0: [1],       # Leaf
    1: [0, 2],    # Internal node
    2: [1, 3],    # Internal node
    3: [2]        # Leaf
}
count_leaves(tree)
```

```python
def is_binary_tree(graph):
    if not graph:
        return True
    
    for node in graph:
        if len(graph[node]) > 2:  # Check if node has > 2 children
            return False
    
    return True
tree = {
    0: [1, 2],   # Root (2 children)
    1: [],        # Leaf (0 children)
    2: [3, 4],    # Internal node (2 children)
    3: [],        # Leaf
    4: []         # Leaf
}
is_binary_tree(tree)
```

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
def find_height(node):
    if node is None:
        return -1  # height of null node is -1 (or 0 depending on convention)

    left_childs_height = find_height(node.left)
    right_childs_height= find_height(node.right)

    return 1 + max(left_childs_height, right_childs_height)
# Create a sample binary tree
#         A
#        / \
#       B   C
#      /   / \
#     D   E   F

A = TreeNode("A")
B = TreeNode("B")
C = TreeNode("C")
D = TreeNode("D")
E = TreeNode("E")
F = TreeNode("F")

A.left = B
A.right = C
B.left = D
C.left = E
C.right = F

# Find height of node C
print("Height of node C:", find_height(C))  # Output: 1
def find_depth(root, target, depth=0):
    if root is None:
        return -1  # Not found

    if root == target:
        return depth

    # Search left
    left = find_depth(root.left, target, depth + 1)
    if left != -1:
        return left

    # Search right
    return find_depth(root.right, target, depth + 1)

# Tree
A = TreeNode("A")
B = TreeNode("B")
C = TreeNode("C")
D = TreeNode("D")
E = TreeNode("E")
F = TreeNode("F")

A.left = B
A.right = C
B.left = D
C.left = E
C.right = F

# Compute depth
print("Depth of node A:", find_depth(A, A))  # 0
print("Depth of node B:", find_depth(A, B))  # 1
print("Depth of node D:", find_depth(A, D))  # 2
print("Depth of node F:", find_depth(A, F))  # 2
```
